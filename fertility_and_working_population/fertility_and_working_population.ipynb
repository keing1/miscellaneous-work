It's often been stated that when a population has a low fertility rate, that a smaller number of working age people will need to provide resources for a larger number of older people, and so society as a whole will get poorer. This seems somewhat incomplete because it ignores the fact that populations with lower fertility rates will have fewer kids to support. So how does the population that is working age (by OECD - 15 to 64), change as a function of true fertility rate?

Assumptions:
1. True fertility rate is static over time. (Maybe not accurate for today's societies, but still relevant if any countries manage to find a steady-state TFR < 2.)
2. Women are equally likely to have a kid any year between 20 and 40. I don't think the exact distribution here should matter all that much though.
3. Society has a life expectancy following Germany - no special reason, they are just a middle-of-the-road country in the EU for life expectancy
