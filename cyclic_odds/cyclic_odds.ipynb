```python
max_num = 100

out_list = []
for i in range(1, max_num+1):
    if i == 1:
        out_list.append(1)
    else:
        out_list.append(1/i + (i-1)/i * (1-out_list[(i+1)//2-1]))
```

```python
out_list[9]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(range(1, max_num+1), out_list)
plt.xlabel('Initial number')
plt.ylabel('Probability of losing')
plt.title('Chance of losing for different initial numbers')
```
